```python
import re
import os
import subprocess


# Load HTML file
with open("HTML_files/RG-50.030.0566_trs_en_cleaned.html", "r", encoding="utf-8") as file:
    html_content = file.read()

pattern = r"\s*(Contact reference@ushmm.org for further information about this collection)|(http://collections.ushmm.org)|(https://collections.ushmm.org)\s*"

# Apply regex filtering
filtered_html = re.sub(pattern, "", html_content, flags=re.DOTALL)


# Save the cleaned file
with open("a_filtered_test.html", "w", encoding="utf-8") as file:
    file.write(filtered_html)

try:
    result=subprocess.run(["diff","-u","HTML_files/RG-50.030.0566_trs_en_cleaned.html", "a_filtered_test.html"],text=True,capture_output=True,check=True)
    print(result.stdout)
except subprocess.CalledProcessError as e:
    print(e.stdout)


print("Filtered HTML saved as 'filtered_output.html'")
```
